# 1.Load libraries


```r
options(warn=-1)
library(wbstats)
library(ggplot2)  
library(lubridate)
library(dplyr)
library(ggthemes)
```

# Example of searching variable for population

```r
search_vars <- wbsearch(pattern = "total population")
head(search_vars)
```

# 2. Selected variables for Western Balkans

```r
# Total Population
dataset_population <- wb( indicator = "SP.POP.TOTL",startdate = 1991, enddate = 2020,country = c("MK","BA","AL","SRB","HR","ME","KS"))%>%
dplyr::mutate(value=value/1000000)
p<-ggplot(dataset_population, aes(x = date, y = value, color = country, shape = country, group = country)) +
  geom_point() + 
  geom_line() +
  labs(title = " Total Population in Western Balkans (in millions)") +
  xlab("Year") +
  ylab("Population (in millions)") +
  scale_x_discrete(breaks=seq(1991,2020,5)) 

p + theme_economist() + scale_fill_economist()
```

```r
# GDP 
dataset <- wb(indicator = "NY.GDP.MKTP.CD",startdate = 1996, enddate = 2020,country = c("MK","BA","AL","SRB","HR","ME","KS"))%>%
dplyr::mutate(value=value/1000000)
p<-ggplot(dataset, aes(x = date, y = value, color = country, shape = country, group = country)) +
  geom_point() + 
  geom_line() +
  labs(title = "GDP in Western Balkans in US ") +
  xlab("Year") +
  ylab("US millions ") +
  scale_x_discrete(breaks=seq(1991,2020,5)) 

p + theme_economist() + scale_fill_economist()
```

```r
# Comparision GDP per capita Western Balkans
dataset_gdp <- wb( indicator = "NY.GDP.PCAP.CD",startdate = 1996, enddate = 2020,country = c("MK","BA","AL","SRB","HR","ME","KS"))
p<-ggplot(dataset_gdp, aes(x = date, y = value, color = country, shape = country, group = country)) +
  geom_point() + 
  geom_line() +
  labs(title = " Comparision GDP per capita in Western Balkans ") +
  xlab("Year") +
  ylab("GDP per capita (current US$)") +
  scale_x_discrete(breaks=seq(1996,2020,5)) 

p + theme_economist()+ scale_fill_economist()
```

```r
# Tax payments by businesses are the total number of taxes paid by businesses, including electronic filing
dataset <- wb( indicator="IC.TAX.PAYM",startdate = 1991, enddate = 2020,country = c("MK","BA","AL","SRB","HR","ME","KS"))
p<-ggplot(dataset, aes(x = date, y = value, color = country, shape = country, group = country)) +
  geom_point() + 
  geom_line() +
  labs(title = " Tax payments (number) in Western Balkans ") +
  xlab("Year") +
  ylab("Tax payments (number)") +
  scale_x_discrete(breaks=seq(1991,2020,5)) 
p + theme_economist() + scale_fill_economist()
```

```r
# Profit tax is the amount of taxes on profits paid by the business.
dataset <- wb( indicator = "IC.TAX.PRFT.CP.ZS",startdate = 1991, enddate = 2020,country = c("MK","BA","AL","SRB","HR","ME","KS"))
p<-ggplot(dataset, aes(x = date, y = value, color = country, shape = country, group = country)) +
  geom_point() + 
  geom_line() +
  labs(title = " Profit tax (% of commercial profits) ") +
  xlab("Year") +
  ylab("% of commercial profits") +
  scale_x_discrete(breaks=seq(1991,2020,5)) +
  theme_minimal() +
  theme(panel.grid.minor.x = element_blank())
p + theme_economist() + scale_fill_economist()
```
